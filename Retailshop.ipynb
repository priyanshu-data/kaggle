```python
Customer=pd.read_csv("C:/Users/PRIYA/Desktop/CASE STUDY/RETAIL CUSTOMER DAY TO DAY ANALYSIS/Customer.csv")
Customer.head()
```

```python
Prod_cat=pd.read_csv("C:/Users/PRIYA/Desktop/CASE STUDY/RETAIL CUSTOMER DAY TO DAY ANALYSIS/prod_cat_info.csv")
Prod_cat.head()
```

We can change the Prod_cat to a new table

```python
Prod_cat1=Prod_cat[['prod_cat_code','prod_cat']]
Prod_cat1
```

```python
Prod_cat2=Prod_cat[['prod_sub_cat_code','prod_subcat']]
Prod_cat2
```

```python
transactions=pd.merge(Transactions,Prod_cat1,how= 'outer')
transactions.head()
```

```python
transactions=pd.merge(Transactions,Prod_cat,how='outer')
transactions
```

```python
transactions=pd.merge(transactions,Customer,left_on="cust_id",right_on='customer_Id')
transactions.head()
```

```python
transactions = transactions.drop(['prod_subcat_code', 'prod_cat_code','customer_Id'] , 1) #drop cust_id column
transactions.head()
```

# There are negative values in the (Qty,Rate and total_amt), need to convert it into positive values using .abs()

```python
transactions.Qty=transactions.Qty.abs()
transactions.Rate=transactions.Rate.abs()
transactions.total_amt=transactions.total_amt.abs()
transactions.head()
```

# Age factor also plays an important role in the analysis.

```python
def calculate_age(born):
    born = datetime.strptime(born, "%d-%m-%Y").date()
    today = date.today()
    return today.year - born.year - ((today.month, today.day) < (born.month, born.day))

transactions['Age'] = transactions['DOB'].apply(calculate_age)
transactions.head()
```

# Now let's go for the unique values in the features.

```python
print("# unique values in STore Type: {0}".format(len(transactions['Store_type'].unique().tolist())))
print("# unique values in City: {0}".format(len(transactions['city_code'].unique().tolist())))
print("# unique values in Customer : {0}".format(len(transactions['cust_id'].unique().tolist())))
print("# unique values in Age: {0}".format(len(transactions['Age'].unique().tolist())))
```

```python
labels = transactions['Store_type'].value_counts().index
sizes = transactions['Store_type'].value_counts().values
explode = (0, 0.1, 0.2, 0.1)  # only "explode" the 2nd and 3rd slices (i.e. 'Hogs')

fig1, ax1 = plt.subplots()
ax1.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

ax1.set_title("Retail Shop Store Types")

plt.show()
```
